````python
import tensorflow
from tensorflow import keras
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
````

- ```Data Augmentation holo ekta technique ja training dataset er size barate ebong model er generalization capability improve korte use kora hoy. Ete original images gulo theke notun images create kora hoy by applying various transformations like rotation, zooming, flipping, shifting, etc. Eivabe model training er somoy beshi diverse data peye thake, jar fole overfitting kom hoy ebong model better performance dey unseen data te.```

````python
img = image.load_img('/Users/abutaher/Desktop2/Deep Learning/Code/Augmentation/5.jpg',
                      target_size = (200,200))
````

````python
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2, 
    horizontal_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2
)


# Ekhane bepar ta ki holo seta bujhiye bolchi: ami ekta image load korlam, tarpor ImageDataGenerator diye kichu augmentation parameters define korlam. Ekhon ami oi image ta ke numpy array te convert korbo, karon ImageDataGenerator sudhu numpy array niye kaj kore. Tarpor ami oi array ta ke reshape kore 4D tensor e convert korbo, jate seta model e input hisebe use kora jai. Finally, ami datagen.flow() method diye augmented images generate korbo, ebong prottekta augmented image ke save korbo specified directory te.
````

````python
img = image.img_to_array(img)
````

````python
type(img)
````

````python
img.shape
````

````python
input_batch = img.reshape(1, 200, 200, 3)
````

````python
i = 0

for output in datagen.flow(input_batch , batch_size = 1, save_to_dir = '/Users/abutaher/Desktop2/Deep Learning/Code/Augmentation/pics'):
    i = i + 1

    if i==10:
        break


# Ekhane ami loop chalacchi jekhane datagen.flow() method diye augmented images generate hocche. Prottekta iteration e ami ekta augmented image pabo, jeta ami specified directory te save korbo. Ami i variable diye count korchi koto gulo image save holo, ebong jokhon i = 10 hoye jabe tokhon loop theke ber hoye jabo. Eivabe ami 10 ta augmented image generate korte parbo.
````
