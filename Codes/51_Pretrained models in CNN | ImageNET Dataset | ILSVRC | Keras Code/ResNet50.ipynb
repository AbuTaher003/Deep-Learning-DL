```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
model = ResNet50(weights='imagenet')
```

```python
image_path = '/Users/abutaher/Desktop2/Deep Learning/Code/51_Pretrained models in CNN | ImageNET Dataset | ILSVRC | Keras Code/Chair.jpg'
img = image.load_img(image_path, target_size=(224, 224))

x = image.img_to_array(img)
x = np.expand_dims(x, axis = 0)
x = preprocess_input(x)

# img_to_array diye amra image ke numpy array te convert korechi. karon model e input hote hole image ke array te convert korte hoy.
# ekhane amra expand_dims diye 1 ta dimension add korechi karon model er input shape hocche (batch_size, height, width, channels)
# so that batch_size er jonne amra 1 ta dimension add korechi jate kore amader single image ta o model e input hote pare.
# preprocess_input diye amra input image ke model er jonno suitable format e niye esechi. karon prottek pretrained model er nijer ekta specific preprocessing technique thake.
```

```python
pred = model.predict(x)
print('Predicted: ', decode_predictions(pred, top = 3)[0])

# [0] diye amra first image er prediction niye aschi karon amra ekbar e sudhu 1 ta image diyechi. Aro details e bolle jodi amra ekbar e multiple image diye thaktam tahole [0], [1], [2] diye prottek image er prediction alada kore nite hoto.
# top=3 diye amra bolchi je amra top 3 prediction dekhte chai.
# decode_predictions diye amra prediction gulo ke human readable format e convert korechi.
```
