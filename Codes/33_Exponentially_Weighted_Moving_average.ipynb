````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv("DailyDelhiClimateTest.csv")
df.head()
````

````python
df = df.iloc[:,0:2]
df.head()
````

````python
plt.scatter(df['date'], df['meantemp'], color = 'red')
plt.show()
````

````python
x1 = df['meantemp'].ewm(alpha = 0.9).mean()
x1
````

- ```Here alpha means Beta```
- ```ewm diye amra Exponentially Weighted moving average ber korechi for each of the rows```

````python
df['EWMA'] = x1
df.head()
````

````python
plt.scatter(df['date'], df["meantemp"], color = 'red')
df['meantemp'].ewm(alpha = 0.9).mean().plot(color = 'blue', label = 'EWMA alpha=0.9')

# plt.plot(df['date'], x1, color = 'black', label = 'EWMA alpha=0.9')
# evabeo EWMA er plot aka jay.


plt.legend()
plt.show()
````
